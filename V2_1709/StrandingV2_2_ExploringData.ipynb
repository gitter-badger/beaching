```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sb
import os
from matplotlib import rcParams
```

```python
sb.reset_defaults()
sb.set_style(style='darkgrid')
```

```python
rcParams['font.size'] = 14
rcParams['xtick.labelsize'] = 14
rcParams['ytick.labelsize'] = 14
rcParams['figure.figsize'] = (10, 6)
```

```python
%matplotlib inline
```

```python
dataDirMain = '/home/madhatter106/DATA/Beaching_Project/CapeCodData/PklJar/'
fp = os.path.join(dataDirMain, 'dfStrand.pkl')
```

```python
df = pd.read_pickle(fp)
```

```python
df.head()
```

```python
df.REPORT_DATE.dt.year.value_counts()
```

```python
df.SPECIES.value_counts(normalize=True)
```

```python
dfLageno = df.loc[df.SPECIES=='Lagenorhynchus acutus']
dfDdelphis = df.loc[df.SPECIES=='Delphinus delphis']
dfDdelphis.to_pickle(os.path.join(dataDirMain, 'dfDdelphisStrand.pkl'))
dfLageno.to_pickle(os.path.join(dataDirMain, 'dfLagenoStrand.pkl'))
```

```python
f, ax = pl.subplots(figsize=(14,6))
df.REPORT_DATE.dt.year.value_counts().sort_index().plot(ls='', marker='.', markersize=15, alpha=0.3,
                                                        ax=ax, label='All', color='k');
dfLageno.REPORT_DATE.dt.year.value_counts().sort_index().plot(ls='', marker='o', color='r', ax=ax,
                                                             label='L. acutus', alpha=0.5);
dfDdelphis.REPORT_DATE.dt.year.value_counts().sort_index().plot(ls='', marker='^', color='g', ax=ax,
                                                               label='D. delphis');
ax.set_xlim((1998, 2015))
ax.set_xticks(np.arange(1999, 2015))

ax.legend();
```
